```python

# ===================== AGENTES =====================
from autogen import ConversableAgent

llm_config = {
    "model": "gpt-4o-mini",      # Modelo leve e rÃ¡pido
    "api_key": os.getenv("OPENAI_API_KEY")
}

planner = ConversableAgent(
    name="PLANNER",
    system_message="""
    VocÃª Ã© o PLANNER.
    Sua funÃ§Ã£o Ã© receber a solicitaÃ§Ã£o e criar um PLANO passo a passo.
    O plano deve ser claro, numerado, e organizado.
    """,
    llm_config=llm_config
)

executor = ConversableAgent(
    name="EXECUTOR",
    system_message="""
    VocÃª Ã© o EXECUTOR.
    Sua funÃ§Ã£o Ã© pegar o plano e transformÃ¡-lo em:
    - Checklist operacional
    - Tarefas com responsÃ¡vel
    - Prazos sugeridos
    Seja direto e objetivo.
    """,
    llm_config=llm_config
)

# ===================== ENTRADA DO USUÃRIO =====================
usuario = "Maria"
user_request = """
Planejar uma reuniÃ£o de definiÃ§Ã£o da sprint do Projeto XPTO.
Objetivo da reuniÃ£o: definir 5 tarefas prioritÃ¡rias, responsÃ¡veis e prazo final.
Data ideal: 15/12/2025
Participantes: Ana (frontend), Paulo (backend), Marcos (QA), JÃºlia (Scrum Master)
Produza: Plano -> Checklist -> ResponsÃ¡veis.
"""

# ===================== CONVERSAÃ‡ÃƒO ENTRE AGENTES =====================
print("ðŸ’¡ Enviando solicitaÃ§Ã£o ao PLANNER...\n")
planner_response = planner.generate_reply(messages=[{"role":"user","content": user_request}])
print("ðŸ”¹ Plano Criado:\n", planner_response, "\n")

print("âš™ Enviando o plano ao EXECUTOR...\n")
executor_response = executor.generate_reply(messages=[{"role":"user","content": planner_response}])
print("ðŸ”¹ Checklist do EXECUTOR:\n", executor_response)

# ===================== SALVANDO EM ARQUIVO =====================
arquivo = "resultado_reuniao.txt"
with open(arquivo, "w", encoding="utf-8") as f:
    f.write("===== PLANNER - PLANO =====\n")
    f.write(planner_response)
    f.write("\n\n===== EXECUTOR - CHECKLIST =====\n")
    f.write(executor_response)

print("\nðŸ“„ Arquivo 'resultado_reuniao.txt' gerado com sucesso!")

# ===================== DOWNLOAD DO ARQUIVO =====================
from google.colab import files
files.download(arquivo)
```
