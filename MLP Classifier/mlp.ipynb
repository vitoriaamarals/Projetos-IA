```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris, load_wine
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import pickle
import warnings

# Ignorar avisos de convergência para manter a saída limpa
warnings.filterwarnings('ignore', category=UserWarning)
```

```python
print("\nAnálise do Dataset Iris")
iris = load_iris()
X_iris, y_iris = iris.data, iris.target
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.3, stratify=y_iris, random_state=42)
print("Dados Iris carregados e divididos.")

# Treinamento
clf_iris = MLPClassifier(random_state=1, max_iter=300).fit(X_train_iris, y_train_iris)
print("Modelo MLP treinado com dados Iris.")

# Salvando e carregando o modelo treinado
filename_iris = 'mlp_model_iris.pkl'
with open(filename_iris, 'wb') as file:
    pickle.dump(clf_iris, file)
print(f"Modelo Iris salvo como '{filename_iris}'")

with open(filename_iris, 'rb') as file:
    loaded_mlp_iris = pickle.load(file)
print(f"Modelo '{filename_iris}' carregado do disco.")

# Predições com o modelo carregado
y_pred_iris = loaded_mlp_iris.predict(X_test_iris)
print("Predições realizadas com o modelo carregado.")

# avaliação
accuracy_iris = accuracy_score(y_test_iris, y_pred_iris)
precision_iris = precision_score(y_test_iris, y_pred_iris, average='weighted')
recall_iris = recall_score(y_test_iris, y_pred_iris, average='weighted')

print(f"\nMétricas de Avaliação (Iris)")
print(f"Acurácia: {accuracy_iris:.4f}")
print(f"Precisão: {precision_iris:.4f}")
print(f"Revocação (Recall): {recall_iris:.4f}")

# Matriz de confusão
cm_iris = confusion_matrix(y_test_iris, y_pred_iris)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_iris, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predito')
plt.ylabel('Verdadeiro')
plt.title('Matriz de Confusão - Dataset Iris')
plt.show()
```

```python
print("\nAnálise do Dataset Wine")
wine = load_wine()
X_wine, y_wine = wine.data, wine.target
X_train_wine, X_test_wine, y_train_wine, y_test_wine = train_test_split(X_wine, y_wine, test_size=0.3, stratify=y_wine, random_state=42)
print("Dados Wine carregados e divididos.")

# Treinamento
clf_wine = MLPClassifier(random_state=1, max_iter=500).fit(X_train_wine, y_train_wine)
print("Modelo MLP treinado com dados Wine.")

# Salvando e carregando o modelo treinado
filename_wine = 'mlp_model_wine.pkl'
with open(filename_wine, 'wb') as file:
    pickle.dump(clf_wine, file)
print(f"Modelo Wine salvo como '{filename_wine}'")

with open(filename_wine, 'rb') as file:
    loaded_mlp_wine = pickle.load(file)
print(f"Modelo '{filename_wine}' carregado do disco.")

# Predições com o modelo carregado**
y_pred_wine = loaded_mlp_wine.predict(X_test_wine)
print("Predições realizadas com o modelo carregado.")

# avaliação
accuracy_wine = accuracy_score(y_test_wine, y_pred_wine)
precision_wine = precision_score(y_test_wine, y_pred_wine, average='weighted')
recall_wine = recall_score(y_test_wine, y_pred_wine, average='weighted')

print(f"\nMétricas de Avaliação (Wine)")
print(f"Acurácia: {accuracy_wine:.4f}")
print(f"Precisão: {precision_wine:.4f}")
print(f"Revocação (Recall): {recall_wine:.4f}")

# Matriz de confusão
cm_wine = confusion_matrix(y_test_wine, y_pred_wine)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_wine, annot=True, fmt='d', cmap='Oranges', xticklabels=wine.target_names, yticklabels=wine.target_names)
plt.xlabel('Predito')
plt.ylabel('Verdadeiro')
plt.title('Matriz de Confusão - Dataset Wine')
plt.show()

print("\nAnálise completa.")
```
